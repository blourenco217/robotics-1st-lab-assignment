```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Threshold 
img = cv2.imread('images/test_draw_1.png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_, img_thr = cv2.threshold(img_gray, 150, 255, cv2.THRESH_BINARY)
# fig, axs = plt.subplots(1)
# axs.set_title("Thresholded")
plt.imshow(img_gray, aspect="auto", cmap="gray")
```

```python
img = cv2.imread('images/test_draw_1.png')
# Convert to gray-scale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, 3, img.shape[0]/64)
# Draw detected circles
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0, :]:
        # Draw outer circle
        cv2.circle(img, (i[0], i[1]), i[2], (0, 255, 0), 2)
        # Draw inner circle
        cv2.circle(img, (i[0], i[1]), 2, (0, 0, 255), 3)
```

```python
plt.imshow(img, aspect="auto", cmap="gray")
```
